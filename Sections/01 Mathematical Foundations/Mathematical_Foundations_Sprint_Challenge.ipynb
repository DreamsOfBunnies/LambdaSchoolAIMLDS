### 2. link a function with a table of values

```
x = [[0,1,2,3,4],[-6.28,-3.14,0,3.14,6.28],[0,0,0,0,0]]

# plot each function in this table
x_1 = np.linspace(1,5,5)
for func in x:
  plt.plot(x_1,func)
  plt.plot(x_1,func, 'bp')
  plt.xlim([1,5])
  
# Are the functions related?  
# The functions are all linear transformations of each other
```

These functions are linear transformations of each other that have been scaled and/or rotated about a point.

### 5. add two vectors together

$z = \begin{pmatrix} 2 \\ -1 \end{pmatrix} + \begin{pmatrix} 2 \\ -1 \end{pmatrix}$

```
x = np.array([2, -1], dtype=int)
x.shape = (2,1)
print('x =\n',x)

y = np.array([2, -1], dtype=int)
y.shape = (2,1)
print('\ny =\n',y)

z = np.add(x, y)
print('\nx + y =\n',z)
```

### 6. subtract a vector from another vector

$z = \begin{pmatrix} 1 \\ 3 \end{pmatrix} - \begin{pmatrix} -2 \\ -3.5 \end{pmatrix}$

```
x = np.array([1, 3], dtype=int)
x.shape = (2,1)
print('x =\n',x)

y = np.array([-2, -3.5], dtype=int)
y.shape = (2,1)
print('\ny =\n',y)

z = np.subtract(x, y)
print('\nx - y =\n',z)
```

### 7. calculate the magnitude of a vector

$|m| = \sqrt{\sum{v_i^2}}$

```
v = [1,2,3,4,5]

# what's the magnitude of v?
mag_v = np.linalg.norm(v)

print(mag_v)
# magnitude is the root of the sum of the squares
```
